```r
library(tidyverse)
```

```r
x = seq(1, 20)
```

```r
x
```

```r
beta_0 = 2
beta_1 = 1.8
```

```r
y =  beta_0 + beta_1 * x 
```

```r
data <- data.frame(x, y)
```

```r
ggplot(data) + 
  geom_point(aes(x=x, y=y))
```

```r
length(x)
```

```r
error = rnorm(length(x), 0, 10)
```

```r
y =  beta_0 + beta_1 * x + error
```

```r
data = data.frame(x, y)
```

```r
ggplot(data) + 
  geom_point(aes(x=x, y=y))
```

```r
model <- lm(y ~ x, data)
summary(model)
```

```r
y_model =  -0.7615 + 1.91 * x
```

```r
ggplot(data) + 
  geom_point(aes(x, y_model), color= "red") + 
  geom_point(aes(x, y), color="black")
```
