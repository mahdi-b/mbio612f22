### Logistic Regrssion

* Linear regression models assume response is quantitative but in many situations, the response variable is instead qualitative
  * A qualitative outcome is one that takes values in a set of possible categories 
* Logistic regression is used when the outcome is qualitative (or categorical)
  * Healthy vs. diseased coral
  * Differentially expressed gene vs. no difference in expression
  * native vs. non-native species
  * etc.

* Logistic Regression is one of many models used in classification


### Classification

* Classification takes a set of inputs and assigns them to a qualitative outcome 
 
* Often, we are interested in predicting the probability that an input belongs to each category
  * We can then take the assignment with the highest probability
  * I.e., more relevant to say the probability of a bleaching event is 0.85, than just the model predicts a bleaching event

### Classes of Logistic Regression

* There are two types of logistic regression:
   1. Simple or binary logistic regression
     * Output consists of two classes

   2. Multiclass logistic regression
    * The outcome can include two or more classes
    * E.g. using data from animal-borne tags and sensors to classify behavior as resting, feeding, playing and aggression.
* We focus on (binary) logistic regression
  * Easy to generalize to more than two classes

### (some) Drawbacks of Using Linear Regression

* The regression line is sensitive to the data
  * outliers can drastically change the intersection point of the threshold on the line
  
* Predicted value can be much smaller than 0 or much larger than 1
  * Impossible to interpret as a probability      

### Logistic Regression

* A very simple yet popular method for modeling qualitative data
  * Popular for classification tasks

* Constrains the predicted value to be between 0 and 1
  * Can be interpreted as a probability for a class
 
* Uses the logistic, or sigmoid function, to model the separation between the classes


$$ g(x) = \frac{1}{1+e^-(\beta_0 + \beta_1~x)}$$


* The logistic function g(x) provides $prob(y=1~|~x, \beta_1, \beta_2)$


### Logistic Regression -- Example

* As with the linear regression, we assume that the parameters of the model $\beta_0$ and $\beta_1$ are unknown.
  * Modeling consists in finding the parameters that best fit the data

```r
beta_0 = 0 
beta_1 = 1
x_axis = seq(-10, 10, 0.01)
y_axis = 1 / (1+exp(-(beta_0 + beta_1 * x_axis)))
```

```r
options(repr.plot.width=14)
ggplot()+
  geom_line(aes(x_axis, y_axis))
```

### Impact of the parameters $\beta_0$ and $\beta_1$

* Like linear regression the parameters have an interpretation
 * $beta_0$ determines where sigmoid function intersection $p=0.5$
 * $beta_1$ determines the "slope" of the sigmoid

### Fitting the Model: Linear Regression Parallel

* For the linear regression, we used the data to parameterize the model
  * We trained the model to find the line that minimizes the RSS
 
  * I.e., among all possible lines, the one selected has the smallest sum of squared residuals      

### Fitting the Model

* In binary logistic regression, the outcomes are 0 and 1, and the predictions are also either 0 or 1

* One trivial approach would consist in identifying the sigmoid function that minimizes the mispredictions
* Using the same method as in linear regression, we could describe the function as:

$$
\frac{1}{n} \sum_1^n (y - g(x))
$$

* Unfortunately, this function is not convex 

  * Does not guarantee that the optimization function will converge to the global minimum
  * This is simply an O=optimizing inconvenience that does not let us find the parameters in a reasonable time

### Fitting the Model- Cont'd


* Instead, given some tuple ($\beta_0$ and $\beta_1$), the following convex function is commonly used to compute the cost for a single point $x$


$$
    cost(g(x),y) = \left\{
    \begin{array}{ll}
        -log(g(x)) & \mbox{ if }  y = 1 \\
        -log(1 - g(x)) & \mbox{ if }  y = 0 \\
    \end{array}
\right.
$$

* Remember that $g(x)$ is simply the probability that the class is 1


### Cost Function Explained - $y=1$

$$
        -log(g(x)) \mbox{ if } y = 1
$$

* For $y = 1$
  - If the prediction is also 1, then we incur no cost 
    * log(1) = 0
  - If the prediction is 0, then we are incurring an infinite cost 
      * $-log(0) \to +\infty$

```r
x_axis = seq(0.000001, 1, 0.01)
y_axis = -log(x_axis)
ggplot()+
   geom_line(aes(x_axis, y_axis)) +
   xlim(0,1) +
   ylim(0,5)
```

### Cost Function Explained - $y=0$

$$
        -log(1 - g_(x)) \mbox{ if }  y = 0 
$$
- for $y = 0$
  - If the prediction is also 0, then we incur no cost 
    * $log(1) = 0$
  - If the prediction is 1, then we are incurring an infinite cost 
    * $-log(0) \to +\infty$




```r
x_axis = seq(0.000001, 1, 0.01)
y_axis = -log(1-x_axis)
ggplot()+
   geom_line(aes(x_axis, y_axis)) +
   xlim(0,1) +
   ylim(0,5)
```

### Total cost

* Therefore, by averaging the cost over all instances in the data, we end up with the parameters ($\beta$s) that best fit the data

* We need to pick the tuple $\beta$ of params which minimizes:

$$
\sum_{\forall x}cost(g(x)_{\beta},y)
$$

* In reality, the cost is calculated slightly differently, but the principle still stands


### Prediction

* Use the parameters $\beta$ identified in the training to predict the probability of true using
$$
p(y=1| x, \beta_0,\beta_1) = g(x) = \frac{1}{1+e^{\beta_0+\beta_1~x}}
$$
 * Probability of False is merely $1- p(y=1| x, \beta_0,\beta_1)$




```r
logistic_reg_model = glm(data=data, bleached~temp, family="binomial")
summary(logistic_reg_model)
```

```r
logistic_reg_model$coefficients[1]
```

```r
beta_0 = logistic_reg_model$coefficients[1]
beta_1 = logistic_reg_model$coefficients[2]


x_axis = seq(min(data$temp)-3, max(data$temp)+3, 0.05)
g_x = 1 / (1+exp(-(beta_0 + beta_1 * x_axis)))
ggplot()+ 
  geom_point(aes(x=data$temp, y=numeric_bleached)) + 
  geom_line(aes(x_axis, g_x)) +
  xlab("Tempe|rature") +
  ylab("Status") +
  theme(text = element_text(size = 22))   
```

### Model Interpretation - Balance


$$
    P(default=1| x, \beta_0, \beta_1) \approx \frac{1}{1 + e^{21.8 - 0.93 ~ temperature}}
$$

- Note that $\beta_1$ reflects the units, which in this case is temprature

- Therefore, an increase of temp by 1 degree, increases the probability of bleachingby 
  
$$
    P(default=1| x=1, \beta_0=-21.8, \beta_1=0.93) \approx \frac{1}{1 + e^{21.8 - 0.93 * 1}} = ?
$$



### Relationship Between Logistic and Linear Model

We have:

$$p(x) = \frac{1}{1+e^-(\beta_0 + \beta_1~x)}$$

or:

$$p(x) = \frac{e^{(\beta_0 + \beta_1~x)}}{1+e^{e(\beta_0 + \beta_1~x)}}$$

This can be written as

$$\frac{p(x)}{1-p(x)} = e^{(\beta_0 + \beta_1~x)}$$



### Relationship Between Logistic and Linear Model - Cont'd

We have that:
    
$$\frac{p(x)}{1-p(x)} = e^{(\beta_0 + \beta_1~x)}$$

By taking log on both sides, we have:

This can be written as

$$log(\frac{p(x)}{1-p(x)}) =\beta_0 + \beta_1~x$$

* Left-hand side is called `log-odds` or `logit`

* Therefore, the linear model has a logit that is linear in $x$

* As we saw above, unlike in the linear model, $\beta_1$ does not correspond to the change associated with a one-unit increase in $x$.

* The `glm` above can be written as:
`logistic_reg_model = glm(data=data, bleached~temp, family=binomial(link='logit'))

```r
logistic_reg_model = glm(data=data, bleached~temp, family=binomial(link='logit'))
summary(logistic_reg_model)
```
